```python
# Import dependencies
import pandas as pd
from google.colab import files
```

```python
# Load in the Migration_Population Dataset
path = "https://github.com/pavlarsen/Final_Project/blob/main/Resources/Data/Migrant_Population_2.xlsx?raw=true"
migrant_population_df = pd.read_excel(path, sheet_name = 1, skiprows = [0, 1,2,3,4,5,6,7,8], header = 1)
migrant_population_df.head(10)
```

```python
# Rename the columns we will use
migrant_population_df.rename(columns={"Region, development group, country or area": "Country_Name", "Location code": "Country_ID", "Total":"Total_Migrant_Population", "Total.1":"Migrant_Male_Population","Total.2":"Migrant_Female_Population"}, inplace = True)
```

```python
# Check the previous code
migrant_population_df.head()
```

```python
# Create a reduced_df that only has the columns we will use
reduced_df = migrant_population_df[["Country_ID", "Country_Name", "Year", "Total_Migrant_Population", "Migrant_Male_Population", "Migrant_Female_Population"]]
reduced_df.head()
```

```python
# Get a list of country codes in order to drop the regions that will not be part of the dataset
code_path = "https://raw.githubusercontent.com/lukes/ISO-3166-Countries-with-Regional-Codes/master/all/all.csv"
code_df = pd.read_csv(code_path, encoding = "UTF-8", header = 0)
country_code = code_df["country-code"].values.tolist()
country_code
```

```python
# Drop the columns that refer to regions (we only want the data for countries)
country_migration_df = reduced_df[migrant_population_df["Country_ID"].isin(country_code)]
country_migration_df.head()
```

```python
# Join the tables and drop the columns that won't be used
merged_df = pd.merge(country_migration_df, code_df, left_on = "Country_ID", right_on = "country-code", how="left")
country_code_df = merged_df[["Country_ID", "alpha-2", "Country_Name", "Year", "Total_Migrant_Population", "Migrant_Male_Population", "Migrant_Female_Population"]]
```

```python
# Check the DataFrame
country_code_df.head()
```

```python
# Change the name column for the "alpha-2" column
renamed_df = country_code_df.rename(columns={"alpha-2":"Country_ID_Alpha"})
```

```python
# Check for null values()
renamed_df.isna().sum()
```

```python
# Identify the rows that have NaN values.
renamed_df[renamed_df["Country_ID_Alpha"].isna()]
```

```python
# Replace the nan values with the country code for Namibia ("NA")
complete_df = renamed_df.fillna("NA")
complete_df.isna().sum()
```

```python
# Check that the dataset has the correct data types
complete_df.dtypes
```

```python
complete_df.head()
```

```python
# Count rows
complete_df["Country_ID"].count()
```

```python
# Check for empty values
complete_df.loc[complete_df["Migrant_Male_Population"] == 0]
```

```python
# Count the empty values
complete_df.loc[(complete_df["Migrant_Male_Population"] == 0) | (complete_df["Migrant_Female_Population"] == 0) | (complete_df["Total_Migrant_Population"] == 0)].count()
```

```python
# Make a copy to eliminate countries with empty data
complete_final = complete_df.copy()
```

```python
# Delete all rows with columns 'Migrant_Male_Population',"Migrant_Female_Population", "Total_Migrant_Population" that have value of 0
index_empty = complete_final.loc[(complete_final["Migrant_Male_Population"] == 0) | (complete_final["Migrant_Female_Population"] == 0) | (complete_final["Total_Migrant_Population"] == 0)].index
complete_final.drop(index_empty, inplace=True)
complete_final
```

```python
complete_final.loc[(complete_final["Migrant_Male_Population"] == 0) | (complete_final["Migrant_Female_Population"] == 0) | (complete_final["Total_Migrant_Population"] == 0)].count()
```

```python
# Export the DataFrame
complete_final.to_csv("Migrant_Population_Final.csv", index=False)
files.download("Migrant_Population_Final.csv")
```
