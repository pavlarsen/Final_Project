```python
# Import dependencies
import pandas as pd
import numpy as np
from google.colab import files
```

```python
# Load in the Country_Data Dataset
path = "https://raw.githubusercontent.com/pavlarsen/Final_Project/main/Resources/Data/Country_Data.csv"
country_data_df = pd.read_csv(path)
country_data_df.head(10)
```

```python
# Rename the columns we will use
country_data_df.rename(columns={"GENC": "Country_ID_Alpha", "Country/Area Name": "Country_Name", "Area in Square Kilometers":"Country_Area", "Population":"Total_Country_Population","Annual Growth Rate %":"Annual_Growth_Rate", "Population Density (People per Sq. Km.)":"Population_Density", "Total Fertility Rate":"Total_Fertility_Rate", "Crude Birth Rate":"Crude_Birth_Rate", "Life Expectancy at Birth, Both Sexes":"Life_Expectancy_at_Birth", "Infant Mortality Rate, Both Sexes":"Infant_Mortality_Rate", "Crude Death Rate":"Crude_Death_Rate","Net Migration Rate":"Net_Migration_Rate"}, inplace = True)
     
```

```python
# Check the previous code
country_data_df.head()
```

```python
# Reduced DF with main columns
final_columns_df = country_data_df[["Country_ID_Alpha", "Country_Name", "Year", "Total_Country_Population", "Annual_Growth_Rate", "Country_Area", "Population_Density", "Total_Fertility_Rate", "Crude_Birth_Rate", "Life_Expectancy_at_Birth", "Infant_Mortality_Rate", "Crude_Death_Rate", "Net_Migration_Rate"]]
final_columns_df.head()
```

```python
# Add boolean column based on "Net_Migration_Rate"
final_columns_df["Migration_Flag"] = np.where(final_columns_df["Net_Migration_Rate"] >= 0, True, False)
final_columns_df.head()
```

```python
# Check for null values
final_columns_df.isna().sum()
```

```python
# Identify rows with NaN values
final_columns_df[final_columns_df["Country_ID_Alpha"].isna()]
```

```python
# Replace NAN values of country Namibia to "NA"
final_columns_df["Country_ID_Alpha"] = final_columns_df["Country_ID_Alpha"].fillna("NA")
final_columns_df.isna().sum()
```

```python
# Identify rows with NaN values
final_columns_df[final_columns_df["Annual_Growth_Rate"].isna()]
```

```python
# Remove rows with NAN values
final_columns_df = final_columns_df.dropna()
final_columns_df.isna().sum()
```
